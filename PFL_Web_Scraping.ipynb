## 4.1 Stworzenie słownika z odnośnikami

Podczas zapoznawania się ze strukturą strony zauważyłem że w stopce głównej witryny znajdują się hiperłącza odsyłające do poszczególnych lat sezonów:

Postanowiłem, na podstawie tych łaczy zacząć budowanie swojego słownika. Zbadałem strukutrę HTML strony przu pomocy narzędzi deweloperskich przeglądarki. Do pobrania strony został użyty pakiet <b>request</b>. Pozwala on na wysyłanie zapytaniań HTTP POST/GET to serwera i otrzymywanie odpowiedzi. Poruszanie się po selektorach i tagach dokumentu html użyłem pakietu <b>beautifulsoup</b>.

Pobranie odnośników ze stopki zostało wykonane następująco:

```python
from bs4 import BeautifulSoup
import requests as rq

#pobranie strony 
page = rq.get('http://www.hppn.pl/liga')

#stworzenie drzewa html
content = page.content
soup = BeautifulSoup(content, 'html.parser')

#odszukanie linków do poszczególnych sezonów z footera strony
seasons_urls = soup \
.find(id="footerSection") \
.find(id="footerMenu4") \
.find_all("a")

#usunięcie odnośnika "wszystkie sezony"
del seasons_urls[-1]

# wypisanie odnośników do poszczególnych sezonów
for url in seasons_urls:
    print(url.get("href"))
```

Każda ze stron przedstawionych powyżej zostanie otwarta i przefiltrowana po raz kolejny. Tym razem po otwarciu strony dla konkretnego roku, sprawdziłem czy w nawigacji istnieje informacja o I lidze. Jeśli tak, to wyciągnąłem odnośniki do tabeli o resultatów dla konkretnego roku i zapisałem je w słowniku <i>seasons_urls_dict:</i>
<br>
<br>
<figure>
    <img src ="screens/obrazek4.PNG"></img>
    <figcaption style="text-align: center;">Struktura html strony poświęconej konkretnemu sezonowi piłkarskiemu</figcaption>
</figure>
<br>

```python
import json
#pusty słownik na odnośniki zesonów
seasons_urls_dict = {}

#pętla po wyciągniętych odnośnikach z footera
for url in seasons_urls:
    
    #pobranie strony 
    season_page = rq.get(url.get('href'))
    season_content = season_page.content
    season_soup = BeautifulSoup(season_content, 'html.parser')
    
    #sprawdzenie czy istnieje pierwsza liga
    menu_part = season_soup \
    .find(id="leftMenu") \
    .find(id="leftMenuOptions")
    
    first_league_link = menu_part.find_all("a", string="I liga")
    
    # jeśli istnieje I liga pobierz odnośniki 'Tabeli' i 'Rezultatów' i zapis do słownika
    if(len(first_league_link) > 0):
        
        #pobieramy sekcję odpowiedzialną z informacjami o I lidze
        first_league_section = first_league_link[0].parent.parent;
        
        #zapytanie przeszukające nagłówek pierwszego poziomu z celu znalezienia nagłówków 'Tabela' oraz 'Rezultaty'
        first_league_table_href = first_league_section.find("a", string="Tabela")
        first_league_result_href = first_league_section.find("a", string="Rezultaty")
        
        if(len(first_league_table_href) > 0 and len(first_league_result_href) > 0):
            
            #pobranie roku sezonu znajdujacego sie w atrybucie 'title' odnośnika
            title = url.get('title')
            seasons_urls_dict[title] = {
                'Table' : first_league_table_href.get("href"),
                'Results': first_league_result_href.get("href"),
            }

print(len(seasons_urls_dict))
print(json.dumps(seasons_urls_dict, indent=2))
```

Cały słownik zapisałem do dataframe'u dostarczanego przez pakiet <b>pandas</b>:

```python
import pandas as pd

crawl_df = pd.DataFrame(seasons_urls_dict)
crawl_df.head()
```

W taki sposób udało się utworzyć słownika zgodnie z założeniami. Za klucz posłużył rok sezonu. Każdy sezon ma zapisany w słowniku odnośnik do swojej witryny tabela oraz resultaty. W dalszej części raportu zostanie przedstawiona analiza obu witryna dla jednego wybranego sezonu.

## 4.2 Analiza witryny 'tabela'

W tej części raportu zostanie poddana obróbce pojedyncza witryna 'tabela' Tak jak we wcześniejszych krokach strona zostanie pobrana i za pomocą odpowiednich selektorów zostanie wyciągnięta tabelka html z zestawieniem. Do zapisania tabelki html do dataframe'u posłuży pakiet pandas. Pandas ma wygodną metodę read_html pozwalającą od razu zapisać tabelę dokumentu html do dataframe'u. Operacje oraz odpowiednie dostosowanie danych będą wykonywane już na tej strukturze.

```python
#pobranie odnośników ze słownika dla sezonu '2009/10'
df_season_info = crawl_df['sezon 2009/10']
print(df_season_info)
```

```python
#wyciągnięcie informacji ze odnośnika tabela dla sezonu '2009/10'
table_page = rq.get(df_season_info.Table)
table_soup = BeautifulSoup(table_page.content, 'html.parser')

result_table = table_soup \
.find("div", {"class" : "standingsSection"}) \
.find("table", {"class" : "standings linkUnderline"}) 

#wykorzystanie pakietu pandas do sparsowania tabelki html
df = pd.read_html(str(result_table), header=0)

df_table = df[0]

#definicja kolumn dataframeu
df_table.columns = ['Promotion', 'dupId', 'TeamName', 'MatchCount', \
                    'Score', 'Goals', 'WinCount', 'DrawCount', 'LoseCount']
df_table = df_table.drop(['dupId', 'Promotion'], axis=1)
display(df_table)
```

W kolumnie <b>Goals</b> mamy informacje zarówno o ramkach zdobytych jak i straconych przez zespół piłkarski. Postanowiłem kolumnę Goals rozdzielić na dwie odzielne kolumny: <b>GoalsScore</b> oraz <b>GoalsLose</b>:

```python
splited_goals = df_table["Goals"].str.split("-", expand=True)
splited_goals
```

```python
#dodanie nowych kolumn do dataframeu
df_table['GoalsScore'] = splited_goals[0]
df_table['GoalsLose'] = splited_goals[1]

df_table.head()
```

```python
#dodanie kolumny z rokiem sezonu
df_table['Season'] = 'sezon-2008-09'

if('Goals' in df_table.columns.values):
    df_table = df_table.drop('Goals', axis=1)

#zmiana ustawienia kolumn -> kolumna 'Season' przeniesiona na początek
df_columns = df_table.columns.tolist()
new_order_columns =df_columns[-1:] + df_columns[:-1]
df_table = df_table[new_order_columns]

display(df_table)
```

Tak przygotowane dane są już gotowe do zapisu bądź analizy. W dalszej kolejności została stworzona klasa <b>SeasonTableParser</b>. Wyżej przedstawione operacje zostały sprowadzone do funkcji, które umożliwiają ich uruchomienie dla każdego sezonu ze słownika:

```python
class SeasonTableParser:
    season_name = ''
    def __init__(self, season_name):
        self.season_name = season_name
    
    def getSeasonTableDataFrame(self, season_info):
        table_page = rq.get(season_info.Table)
        table_soup = BeautifulSoup(table_page.content, 'html.parser')
        
        result_table = table_soup \
        .find("div", {"class" : "standingsSection"}) \
        .find("table", {"class" : "standings linkUnderline"}) 
        df = pd.read_html(str(result_table), header=0)

        df_table = df[0]
        df_table.columns = ['Promotion', 'dupId', 'TeamName', 'MatchCount', \
                            'Score', 'Goals', 'WinCount', 'DrawCount', 'LoseCount']
        df_table = df_table.drop(['dupId', 'Promotion'], axis=1)
        return df_table
    
    def appendGoalsInformation(self, season_data_frame):
        splited_goals = season_data_frame["Goals"].str.split("-", expand=True)
        season_data_frame['GoalsScore'] = splited_goals[0]
        season_data_frame['GoalsLose'] = splited_goals[1]
        season_data_frame['Season'] = self.season_name
        
        if('Goals' in season_data_frame.columns.values):
            season_data_frame = season_data_frame.drop('Goals', axis=1)

        #reorder columns
        df_columns = season_data_frame.columns.tolist()
        new_order_columns =df_columns[-1:] + df_columns[:-1]
        season_data_frame = season_data_frame[new_order_columns]
        return season_data_frame
    
    def getSeasonTableData(self, season_info):
        season_table = self.getSeasonTableDataFrame(season_info)
        season_table = self.appendGoalsInformation(season_table)
        return season_table
```

Poniżej zostało przeprowadzone testowe uruchomienie parsera dla innego sezonu. DataFrame wygląda identycznie jak poprzedni, wiec parser zadziałał w sposób prawidłowy.

```python
df_test = crawl_df['sezon 2009/10']
tableParser = SeasonTableParser(df_test.name)
df = tableParser.getSeasonTableData(df_test)
display(df)
```

Ostatnim etapem jest uruchomie parsera w pętli dla każdego sezonu. Na tym etapie został utworzony kolejny słownik <i>seasons_tables_dict</i> na dataframe'y tabel dla każdego sezonu. Za klucz w słowniku poraz kolejny posłuży rok sezonu piłkarskiego. Widać poniżej, że parser nie wyrzucił błedu dla żadnego sezonu. Jest to bardzo dobry wynik i możliwe, że wszystkie dane zostały zapisane w sposób prawidłowy.

```python
#słownik na tabele sezonów
seasons_tables_dict = {}

for season in crawl_df:
    try:
        #uruchomienie parsera dla aktualnej iteracji
        data_info = crawl_df[season]
        tableParser = SeasonTableParser(data_info.name)
        df = tableParser.getSeasonTableData(data_info)
        seasons_tables_dict[season] = df
        print("Created table frame for:" + season)
    except:
        print("Error for:" + season)
```

```python
#testowe sprawdzenie dataframe'u z utworzonego słownika
display(seasons_tables_dict['sezon 2011/12'])
```

Słownik <i>seasons_tables_dict</i> został sprowadzony do jednego dużego dataframe'u. Zmienione zostały także wartości w kolumnie <b>Season</b>, z której został pozostawiony tylko rok oraz został usunięty powtarzający sie wyraz 'sezon'

```python
#utworzenie jednego dataframeu ze słownika
concated_frame = pd.concat(seasons_tables_dict.values(), ignore_index=True)

#usunięcie przedrostka 'sezon' z kolumny Season oraz zmiana w wyswietlaniu roku sezonu
concated_frame['Season'] = concated_frame.Season.str.replace('sezon', '').str.replace('/', '-')
display(concated_frame)
```

Ostatnią operacją jest zapisanie całego dataframe'u do pliku csv przy użyciu pakietu <b>pandas</b>:

```python
concated_frame.to_csv('seasons_tables.csv', index=False)
```

## 4.3 Analiza witryny 'rezultaty'

W tej części raportu zostanie poddana obróbce pojedyncza witryna 'rezultaty' Zostały podjęte podobne krok jak przy analizie stron zawierająch tabele.

```python
df_season_info = crawl_df['sezon 2009/10']
print(df_season_info)
```

```python
table_page = rq.get(df_season_info.Results)
table_tree = BeautifulSoup(table_page.content, 'html.parser')
result_table = table_tree \
.find("div", {"class" : "resultsSection"}) \
.find("table", {"class" : "standard"})

df = pd.read_html(str(result_table), header=0)

df_result = df[0]
display(df_result)
```

W przypadku analizy tabeli resultatów zauważyłem, że ma ona bardziej skomplikowaną formę w porównianu do poprzedniego przypadku. Występują wiersze zawierające braki danych oraz w kolumnie data co jakiś czas pokazuje się informacja o numerze kolejki. Postanowiłem wykorzystać tą informację i dodać do odzielnej kolumny o nazwie <b>Queue</b>. Informacja o kolejce pojawia się z interwałem 9. Stworzyłem wiec listę z powtarzającymi się numerami kolejki, którą dodałem do dataframe'u resultatów.

```python
import numpy as np

#pobranie wierszy zawierających wyraz 'kolejka'
queue_properties = df_result.loc[df_result['data'].str.contains("kolejka")]

#pobranie wartości indeksów dla tych wierszy
indexes = queue_properties.index.values

#pobranie ostatniego indexu dataframe'u rezultatów
last_result_index = df_result.index[-1] + 1
indexes = np.append(indexes, last_result_index)

#stworznie listy z informacją o numerze kolejki
queue_result= []
counter = 1
for i in range(len(indexes) - 1):
    queue_list = [counter] * (indexes[i+1] - indexes[i])
    queue_result.extend(queue_list)
    counter= counter + 1

print(queue_result)
```

```python
if('Queue' not in df_result.columns.values):
    df_result["Queue"] = queue_result

df_result = df_result[df_result.mecz.notnull()]
result_columns = ["MatchDate", "Winner", "Result", "Looser", "Queue"]
df_result.columns = result_columns
display(df_result)
```

Na pierwszy rzut oka stwierdziłem, że w drugiej kolumnie zawsze widnieje drużyna, która odniosła zwycieństwo a w czwartej, która odnosła porażkę. Po dokładniejszym przyjrzeniu się na dataframe <i>df_result</i> zauważyłem, że była to błędna obserwacja. Postanowiłem więc odpowednio zodyfikować kolumnę w wynikiem tak, żeby wynik drużyny wygranej zawsze znajdował się po lewej stronie znaku ":".
```python
import numpy as np

class SeasonResultParser:
    season_name = ''
    def __init__(self, season_name):
        self.season_name = season_name
    
    def rearangeResult(self, result):
        resultSplit = result["Result"].split(":")
        winner = result["Winner"]
        looser = result["Looser"]
        if(len(resultSplit) == 2):
            if(resultSplit[0] < resultSplit[1]):
                result["Winner"] = looser
                result["Looser"] = winner
                result["Result"] = resultSplit[1] + ":" + resultSplit[0]
                result["WinnerGoals"] = resultSplit[1]
                result["LooserGoals"] =  resultSplit[0]
            else:
                result["WinnerGoals"] = resultSplit[0]
                result["LooserGoals"] =  resultSplit[1]
            return result
    
    def getSeasonResultDataFrame(self, season_info):
        result_page = rq.get(df_season_info.Results)
        result_tree = BeautifulSoup(result_page.content, 'html.parser')
        result_table = result_tree \
        .find("div", {"class" : "resultsSection"}) \
        .find("table", {"class" : "standard"})

        df = pd.read_html(str(result_table), header=0)

        df_result = df[0]
        df_result.columns = ["MatchDate", "Winner", "Result", "Looser"]
        return df_result
    
    def getQueueList(self, season_data_frame):
        queue_properties = season_data_frame \
        .loc[season_data_frame['MatchDate'].str.contains("kolejka")]
        indexes = queue_properties.index.values
        last_result_index = season_data_frame.index[-1] + 1
        indexes = np.append(indexes, last_result_index)

        queue_result= []
        counter = 1
        for i in range(len(indexes) - 1):
            queue_list = [counter] * (indexes[i+1] - indexes[i])
            queue_result.extend(queue_list)
            counter= counter + 1

        return queue_result
    
    def prepareResultTable(self, season_data_frame, queue_list):
        season_data_frame["Queue"] = queue_list
        season_data_frame["Season"] = self.season_name

         #reorder columns
        df_columns = season_data_frame.columns.tolist()
        new_order_columns =df_columns[-1:] + df_columns[:-1]
        season_data_frame = season_data_frame[new_order_columns]

        season_data_frame = season_data_frame[season_data_frame.Result.notnull()]
        return season_data_frame
    
    def getResultTableData(self, season_info):
        season_table = self.getSeasonResultDataFrame(season_info)
        queue_list = self.getQueueList(season_table)
        season_table = self.prepareResultTable(season_table, queue_list)
        season_table = season_table.apply(self.rearangeResult, axis=1)
        season_table = season_table.dropna(how="any")

        return season_table
```

## 6.1 Utworzenie słownika z adresami tabel i resultatów

```python
from bs4 import BeautifulSoup
import requests as rq
import pandas as pd
import json

#pobranie strony 
page = rq.get('http://www.hppn.pl/liga')

#stworzenie drzewa html
content = page.content
soup = BeautifulSoup(content, 'html.parser')

#odszukanie linków do poszczególnych sezonów z footera strony
seasons_urls = soup \
.find(id="footerSection") \
.find(id="footerMenu4") \
.find_all("a")

#usunięcie odnośnika "wszystkie sezony"
del seasons_urls[-1]

#pusty słownik na odnośniki zesonów
seasons_urls_dict = {}

#pętla po wyciągniętych odnośnikach z footera
for url in seasons_urls:
    
    #pobranie strony 
    season_page = rq.get(url.get('href'))
    season_content = season_page.content
    season_soup = BeautifulSoup(season_content, 'html.parser')
    
    #sprawdzenie czy istnieje pierwsza liga
    menu_part = season_soup \
    .find(id="leftMenu") \
    .find(id="leftMenuOptions")
    
    first_league_link = menu_part.find_all("a", string="I liga")
    
    # jeśli istnieje I liga pobierz odnośniki 'Tabeli' i 'Rezultatów' i zapis do słownika
    if(len(first_league_link) > 0):
        
        #pobieramy sekcję odpowiedzialną z informacjami o I lidze
        first_league_section = first_league_link[0].parent.parent;
        
        #zapytanie przeszukające nagłówek pierwszego poziomu z celu znalezienia nagłówków 'Tabela' oraz 'Rezultaty'
        first_league_table_href = first_league_section.find("a", string="Tabela")
        first_league_result_href = first_league_section.find("a", string="Rezultaty")
        
        if(len(first_league_table_href) > 0 and len(first_league_result_href) > 0):
            
            #pobranie roku sezonu znajdujacego sie w atrybucie 'title' odnośnika
            title = url.get('title')
            seasons_urls_dict[title] = {
                'Table' : first_league_table_href.get("href"),
                'Results': first_league_result_href.get("href"),
            }

print(len(seasons_urls_dict))
print(json.dumps(seasons_urls_dict, indent=2))
crawl_df = pd.DataFrame(seasons_urls_dict)
crawl_df.head()
```

## 6.2 Utworzenie danych wyjściowych z tabelami sezonów

```python
class SeasonTableParser:
    season_name = ''
    def __init__(self, season_name):
        self.season_name = season_name
    
    def getSeasonTableDataFrame(self, season_info):
        table_page = rq.get(season_info.Table)
        table_soup = BeautifulSoup(table_page.content, 'html.parser')
        
        result_table = table_soup \
        .find("div", {"class" : "standingsSection"}) \
        .find("table", {"class" : "standings linkUnderline"}) 
        df = pd.read_html(str(result_table), header=0)

        df_table = df[0]
        df_table.columns = ['Promotion', 'dupId', 'TeamName', 'MatchCount', \
                            'Score', 'Goals', 'WinCount', 'DrawCount', 'LoseCount']
        df_table = df_table.drop(['dupId', 'Promotion'], axis=1)
        return df_table
    
    def appendGoalsInformation(self, season_data_frame):
        splited_goals = season_data_frame["Goals"].str.split("-", expand=True)
        season_data_frame['GoalsScore'] = splited_goals[0]
        season_data_frame['GoalsLose'] = splited_goals[1]
        season_data_frame['Season'] = self.season_name
        
        if('Goals' in season_data_frame.columns.values):
            season_data_frame = season_data_frame.drop('Goals', axis=1)

        #reorder columns
        df_columns = season_data_frame.columns.tolist()
        new_order_columns =df_columns[-1:] + df_columns[:-1]
        season_data_frame = season_data_frame[new_order_columns]
        return season_data_frame
    
    def getSeasonTableData(self, season_info):
        season_table = self.getSeasonTableDataFrame(season_info)
        season_table = self.appendGoalsInformation(season_table)
        return season_table
```

```python
#słownik na tabele sezonów
seasons_tables_dict = {}

for season in crawl_df:
    try:
        #uruchomienie parsera dla aktualnej iteracji
        data_info = crawl_df[season]
        tableParser = SeasonTableParser(data_info.name)
        df = tableParser.getSeasonTableData(data_info)
        seasons_tables_dict[season] = df
        print("Created table frame for:" + season)
    except:
        print("Error for:" + season)

print("Tables dict is created")
```

```python
#utworzenie jednego dataframeu ze słownika
concated_tables_frame = pd.concat(seasons_tables_dict.values(), ignore_index=True)

#usunięcie przedrostka 'sezon' z kolumny Season oraz zmiana w wyswietlaniu roku sezonu
concated_tables_frame['Season'] = concated_tables_frame.Season.str.replace('sezon', '').str.replace('/', '-')

concated_tables_frame.to_csv('seasons_tables.csv', index=False)

print("seasons_tables.csv created")
```
